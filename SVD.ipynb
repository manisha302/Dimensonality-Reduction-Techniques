now find out first A.A(TRANSPOSE) which will be needed to calculate right singular eigen vector

```python
def multiply(df):
    df=np.matrix(df)
    df1=df.dot(df.T)
    df1
    return(df1)

df1=multiply(df)
df1
```

```python
#calculating right singular values(V) corresponding to above matrix
```

```python
import scipy.linalg as la
def val2_vect2(df1):
    val2,vect2=la.eig(df1)
    index=val2.argsort()[::-1]
    val2=val2[index]
    vect2=vect2[index]
    return(val2,vect2)

val2,vect2=val2_vect2(df1)
print(val2)
print(vect2)
```

```python
#now we will compute A(TRANSPOSE).A which would be required for left singular vector(U)
```

```python
def multiply1(df):
    df=np.matrix(df)
    df2=(df.T).dot(df)
    df2
    return(df2)

df2=multiply1(df)
df2
```

```python
#calculating left singular vector(V) using corresponding A(TRANSPOSE).A
```

```python
import scipy.linalg as la
def eig_val_eig_vect(df2):
    eig_val,eig_vect=la.eig(df2)
    index=eig_val.argsort()[::-1]
    eig_val=eig_val[index]
    eig_vect=eig_vect[index]
    return(eig_val,eig_vect)

eig_val,eig_vect=eig_val_eig_vect(df2)
print(eig_val)
print(eig_vect)
```

```python
#taking square root of eigen values
```

```python
import numpy as geek
sigma=geek.sqrt(eig_val)
sigma
```

```python
sigma=np.diag(sigma)
sigma=sigma[:10,:10]
sigma
```

```python
def recon_error(sigma,df,eig_vect,vect2,k):
    df=np.array(df)
      
    eig_vect=eig_vect[:,:k]#eigen vector (V)
    eig_vect=eig_vect.T
    
    vect2=vect2[:,:k]#eigen vector(U)
    sigma=sigma[:k,:k]
    rec_A=vect2@sigma
    rec_A=rec_A@eig_vect
    error=sum(sum(df-rec_A)**2)
    return(rec_A,error)
    
res,err=recon_error(sigma,df,eig_vect,vect2,10)
res
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def plot(df,rec_A,k):
    df=np.array(df)
    err=[]
    pc=list(range(1,k))
    for i in range(1,k):
        a,b=recon_error(sigma,df,eig_vect,vect2,i)
        err.append(b)
    plt.plot(pc,err)
    
plot(df,res,5)
```
