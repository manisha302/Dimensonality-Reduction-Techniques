```python
df=df.drop("Unnamed: 0",axis=1)
df
```

```python
mean_df=df.mean()
mean_df
```

```python
std_df=(df-df.mean())/df.std()
std_df.describe()
```

```python
cov_df=np.array(df.cov())
cov_df
```

```python
import scipy.linalg as la
eig_val,eig_vect=la.eig(cov_df)
eig_val,eig_vect
```

```python
np.shape(eig_vect)
```

```python
index=np.argsort(eig_val)
index
```

```python
eig_vect1=eig_vect[index]
eig_vect1=eig_vect1[::-1]
eig_vect1
```

```python
def recon_error(std_df,eig_vect1,k):
    std_df=np.array(std_df)
    G=eig_vect1[0:k]
    x_r=std_df@G.T
    x_n=x_r@G
    error=sum(sum(std_df-x_n)**2)
    return(error)


error=recon_error(std_df,eig_vect1,3)
error
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
def plot(std_df,eig_vect1,k):
    std_df=np.array(std_df)
    err=[]
    pc=list(range(1,k))
    for i in range(1,k):
        err.append(recon_error(std_df,eig_vect1,i))
    plt.plot(pc,err)
```

```python
plot(std_df,eig_vect1,5)
```
